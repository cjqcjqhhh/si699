## GithubRepo Text Preprocess

This piece of code aims at some data pre-processing. Here are the main tasks:

+ Proprocess the `readme` attribute in `GithubRepo` object:
    
    + Recursively remove link
    
    + Remove all html content
    
    + Parse markdown text
    
    + Organized the Filter into a class (not yet done, currently some functions)

```python
instance = github_dataset[3]
s = instance['readme'] # this is a sample readme text
instance["link"]
```

```python
def filter(s):
    # s: the raw markdown string
    while True:
        temp = re.sub(r"\[(.*)\]\(.*\)", r"\1", s)
        if temp == s:
            break
        s = temp # recursively remove link
    s = re.sub('<[^<]+?>', '', s) # remove all html tags
    # s = re.sub("\n|\*|-", "", s) # remove addtional non-English char
    return '\n\n' + s

s = filter(s)
```

```python
def parse_heading(level, s):
    # level: the nth level heading, should be a postive integer
    # s: the filtered md string
    find_pattern = r"\n\n#{%s} (.*)\n\n" % str(level)
    split_pattern = r"\n\n#{%s} .*\n\n" % str(level)
    headings = re.findall(find_pattern, s)
    contents = re.split(split_pattern, s)
    return headings, contents
```
