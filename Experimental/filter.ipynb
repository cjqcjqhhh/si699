## GithubRepo String Filter

This piece of code aims at some data pre-processing. Here are the main tasks:

+ Filter the non-English characters in `readme` attribute in `GithubRepo` object.

```python
import json 
from stackapi import StackAPI
from magi_dataset import *
from tqdm import tqdm
import time
import os

# github dataset with 5000 repo
with open("./ghv9-2.json", encoding='utf-8') as data_file:
    github_dataset = json.loads(data_file.read())
```

```python
instance = github_dataset[8]
instance["link"]
```

```python
import re

s = instance['readme']

s = re.sub(r"\[(.*)\]\(.*\)", r"\1", s) # replace link
# s = re.sub(r"<*>(*)</*>", r"\1", s)
# s = re.sub("\n|\*|-", "", s)
s
```
